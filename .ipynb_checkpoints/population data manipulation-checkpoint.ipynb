```python
import pandas as pd
```

```python
pop = pd.read_csv('population.csv')
pop
```

```python
df = pd.read_excel('enroll.xlsx')
df
```

```python
countries_pop = list(pop['Country (or dependency)'].unique())
len(countries_pop)
```

```python
countries_enroll = list(df['country'].unique())
len(countries_enroll)
```

```python
common_elements = set([c.lower() for c in countries_pop]).intersection([c.lower() for c in countries_enroll])
len(common_elements)
```

```python
non_common_elements = set([c.lower() for c in countries_pop]).difference([c.lower() for c in countries_enroll])
len(non_common_elements)
```
